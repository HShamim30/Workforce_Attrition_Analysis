```python
df = pd.read_csv("../data/Palo_Alto_Networks.csv")
```

```python
df.shape
```

```python
print(df['Attrition'].value_counts())
```

```python
total_emp = len(df)
print(total_emp)
```

```python
exited = df['Attrition'].sum()
print(exited)
```

```python
retained = total_emp - exited
print(retained)
```

```python
attrition_rate = (exited/total_emp)*100
print(attrition_rate)
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Attrition', data=df)
plt.title("Attrition Distribution")
plt.show()
```

```python
dept = df.groupby('Department')['Attrition'].mean()*100
print(dept)
```

```python
dept.sort_values().plot(
    kind='barh',
    figsize=(8,5)
)

plt.xlabel("Attrition Rate (%)")

plt.show()
```

```python
role = df.groupby('JobRole')['Attrition'].mean()*100
print(role)
```

```python
role.sort_values().plot(
    kind='barh',
    figsize=(10,6)
)

plt.show()
```

```python
bins = [18,25,35,45,55,65]

labels = ['18-25','26-35',
          '36-45','46-55',
          '56-65']

df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels
)
```

```python
age = df.groupby('AgeGroup')['Attrition'].mean()*100

print(age)
```

```python
age.plot(kind='bar')
plt.ylabel("Attrition Rate (%)")
plt.show()
```

```python
gender = df.groupby('Gender')['Attrition'].mean()*100
print(gender)
gender.plot(kind='bar')
plt.show()
```

```python
marital = df.groupby('MaritalStatus')['Attrition'].mean()*100
print(marital)
marital.plot(kind='bar')
plt.show()
```

```python
education = df.groupby('Education')['Attrition'].mean()*100
education.plot(kind='bar')
plt.show()
```

```python
edu_field = df.groupby('EducationField')['Attrition'].mean()*100
edu_field.sort_values().plot(
    kind='barh',
    figsize=(8,5)
)
plt.show()
```

```python
bins = [0,2,5,10,20,40]
labels = ['0-2','3-5',
          '6-10','11-20',
          '20+']
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins=bins,
    labels=labels
)
```

```python
tenure = df.groupby('TenureGroup')['Attrition'].mean()*100
tenure.plot(kind='bar')
plt.show()
```

```python
early = df[df['YearsAtCompany']<=2]
rate = early['Attrition'].mean()*100
print(rate)
```

```python
promo = df.groupby(
'YearsSinceLastPromotion'
)['Attrition'].mean()*100
promo.plot()
plt.show()
```

```python
overtime = df.groupby(
'OverTime'
)['Attrition'].mean()*100
print(overtime)
overtime.plot(kind='bar')
plt.show()
```

```python
travel = df.groupby(
'BusinessTravel'
)['Attrition'].mean()*100
print(travel)
travel.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(
    x='Attrition',
    y='DistanceFromHome',
    data=df
)
plt.show()
```

```python
heat = df.pivot_table(
    index='Department',
    columns='JobRole',
    values='Attrition',
    aggfunc='mean'
)

plt.figure(figsize=(15,6))

sns.heatmap(
    heat,
    annot=True,
    cmap='Reds',
    fmt='.2f'
)

plt.title(
'Department vs Job Role Attrition Heatmap'
)

plt.show()
```

```python
plt.figure(figsize=(16,10))

sns.heatmap(
    df.corr(numeric_only=True),
    cmap='coolwarm'
)

plt.show()
```
